```python
import pandas as pd
from sqlalchemy import create_engine
```

# Read 1st csv into DataFrame

```python
csv_file = "./lending_club_csvs/LoanStats0.csv"
loan_df = pd.read_csv(csv_file)
loan_df.head()
```

# Combine all 12 csvs


```python
# Create a place to store all the .csv files together
mega_list = []
for i in range (12): #12
    
    
    csv_file = f"./lending_club_csvs/LoanStats{i}.csv"
    loan_df = pd.read_csv(csv_file)
    mega_list.append(loan_df)
mega_df=pd.concat(mega_list)
```

```python
# Print new, combined datafile
mega_df
```

# Create new data with select columns

```python
# Create a dataframe with all the chosen columns in the sql schema

new_mega_df = mega_df[['id', 'funded_amnt', 'term', 'int_rate', 'installment']]
new_mega_df.head()
#new_mega_df.count()

# consider adding vars "grade", "home_ownership", "annual_inc",
#   "issue_d", "loan_status", "title", "addr_state", "dti", "earliest_cr_line""fico_range_high",
#   "open_acc", "pub_rec", "pub_rec_bankruptcies", "revol_util", "total_acc", "out_prncp", "total_pymnt",
#   "total_rec_prncp", "recoveries", "collections_12_mths_ex_med", "application_type", "chargeoff_within_12_mths",
#   "hardship_type", "hardship_reason", "debt_settlement_flag"
```

# Clean DataFrame

```python
# Remove rows that are mainly blank

mega_df_drop_blank_rows = new_mega_df.dropna(how="any")
mega_df_drop_blank_rows
```

# Connect to local database

```python
rds_connection_string = "postgres:postgres@localhost:5432/lending_club"

engine = create_engine(f'postgresql://{rds_connection_string}')
```

# Check for tables

```python
engine.table_names()
```

# Use pandas to load csv converted DataFrame into database

```python
mega_df_drop_blank_rows.to_sql(name='lending_club', con=engine, if_exists='append', index=False)
```
